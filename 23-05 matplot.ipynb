```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline


x1 = np.linspace(0, 10, 100)

fig = plt.figure()

plt.plot(x1, np.sin(x1), "-")
plt.plot(x1, np.cos(x1), "--")
plt.show()
```

```python
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x1, np.sin(x1), "-")
plt.subplot(2, 1, 2)
plt.plot(x1, np.cos(x1), "--")
plt.show()
```

```python
plt.plot([1,2,3,4] , marker = "*")
plt.ylabel("numbers")
plt.show()
```

```python
plt.plot([1,2,3,4], [4,3,2,1], marker = "o")
plt.xlabel("x -axis")
plt.ylabel("y -axis")
plt.title("Line plot with markers")
plt.show()
```

```python
x = np.linspace(0, 2, 100)
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')   
plt.ylabel('y label')
plt.title('Simple Plot')
plt.legend()
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], marker = "o")
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y, marker = "o")
plt.xticks(x)
plt.yticks(y)
plt.title("Numbers on the axis matching the markers")
plt.xlabel("x -axis")
plt.ylabel("y -axis")
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
fig, ax = plt.subplots(2)


# Call plot() method on the appropriate object
ax[0].plot(x1, np.sin(x1), 'b-')
ax[1].plot(x1, np.cos(x1), 'b-');
plt.show()
```

```python
fig , ax  = plt.subplots(3)

ax[0].plot(x1, np.sin(x1), 'b-')
ax[1].plot(x1, np.cos(x1), 'b-')
ax[2].plot(x1, np.tan(x1), 'b-')
plt.show()
```

```python
fig = plt.figure()

x2 = np.linspace(0, 5 , 10)
y2 = x2**2
axes = fig.add_axes([1, 0.1, 0.8, 1])
axes.plot(x2, y2, 'r-')
axes.set_xlabel('x label')
axes.set_ylabel('y label')  
plt.show()
```

```python
x3 = np.arange(0.0, 6.0, 0.01) 

plt.plot(x3, [xi**2 for xi in x3], 'b-') 

plt.show()
```

```python
x4  = range(1,5)
plt.plot(x4, [xi**1.5 for xi in x4], label ='x^1.5')
plt.plot(x4, [xi**2 for xi in x4], label ='x^2')
plt.plot(x4, [xi**4 for xi in x4], label ='x^4')
plt.plot(x4, [xi**3 for xi in x4], label ='x^3')

plt.legend()
plt.xlabel('x -axis')
plt.ylabel('y -axis')   
plt.title('Line plot with different powers of x')
plt.savefig('my_plot.png')
plt.show()
```
